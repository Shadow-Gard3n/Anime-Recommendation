```python
feature_columns = ['Genres', 'Producers', 'Studios', 'Licensors', 'Rating', 'Source', 'Type', 'Premiered']

for col in feature_columns:
    raw_df[col] = raw_df[col].str.replace(',', ' ', regex=False)

raw_df
```

```python
raw_df['features'] = raw_df['Genres'] + ' ' + raw_df['Producers'] + ' ' + raw_df['Studios'] + ' ' + raw_df['Licensors'] + ' ' + raw_df['Rating'] + ' ' + raw_df['Source'] + ' ' + raw_df['Type'] + ' ' + raw_df['Premiered']
raw_df['features'] = raw_df['features'].str.lower()
raw_df
```

```python
final_df = raw_df[['MAL_ID','Name','features']].copy()
final_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000)
vector = cv.fit_transform(raw_df['features']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
def recommend(anime):
    anime_index = raw_df[raw_df['Name'] == anime].index[0]
    distances = list(enumerate(similarity[anime_index]))
    sorted_distances = sorted(distances, reverse=True, key=lambda x: x[1])
    print(f"Recommendations for: {anime}\n")
    for i in sorted_distances[1:11]:
        anime_name = raw_df.iloc[i[0]]['Name']
        print(anime_name)
```

```python
recommend('Grand Blue')
```
