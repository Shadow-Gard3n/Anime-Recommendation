```python
raw_df['Producers'] = raw_df['Producers'].replace('Unknown', '')
raw_df['Studios'] = raw_df['Studios'].replace('Unknown', '')
raw_df['Licensors'] = raw_df['Licensors'].replace('Unknown', '')
raw_df['Rating'] = raw_df['Rating'].replace('Unknown', '')
raw_df['Source'] = raw_df['Source'].replace('Unknown', '')
raw_df['Type'] = raw_df['Type'].replace('Unknown', '')
raw_df['Premiered'] = raw_df['Premiered'].replace('Unknown', '')
```

```python
raw_df['Producers'] = raw_df['Producers'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Genres'] = raw_df['Genres'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Licensors'] = raw_df['Licensors'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Studios'] = raw_df['Studios'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Source'] = raw_df['Source'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Rating'] = raw_df['Rating'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df['Premiered'] = raw_df['Premiered'].apply(lambda x: x.replace(" ", ""))
```

```python
raw_df
```

```python
feature_columns = ['Genres', 'Producers', 'Studios', 'Licensors', 'Rating', 'Source', 'Type', 'Premiered']

for col in feature_columns:

    raw_df[col] = raw_df[col].apply(lambda x: ','.join([tag.replace(' ', '') for tag in x.split(',')]))
    raw_df[col] = raw_df[col].str.replace(',', ' ')
raw_df
```

```python
raw_df['features'] = (raw_df['Genres'] + ' ') * 3 + raw_df['Producers'] + ' ' + raw_df['Studios'] + ' ' + raw_df['Source'] + ' ' + raw_df['Rating'] + ' ' + raw_df['Type'] + ' ' + raw_df['Premiered']
raw_df['features'] = raw_df['features'].str.lower()
raw_df
```

```python
final_df = raw_df[['MAL_ID','Name','features']].copy()
final_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(max_features=5000)
vector = tfidf.fit_transform(final_df['features']).toarray()
similarity = cosine_similarity(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
def recommend(anime_name):
    try:
        anime_index = raw_df[raw_df['Name'] == anime_name].index[0]
    except IndexError:
        print(f"Error: Anime '{anime_name}' was not found in the dataset.")
        return

    distances = list(enumerate(similarity[anime_index]))
    sorted_distances = sorted(distances, reverse=True, key=lambda x: x[1])

    print(f"Recommendations for: {anime_name}\n")
    for i in sorted_distances[1:11]:
        anime_title = raw_df.iloc[i[0]]['Name']
        print(anime_title)
```

```python
recommend('Grand Blue')
```
